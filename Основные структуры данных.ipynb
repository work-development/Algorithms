```python
def insert_node(node, index, value):
    head = node
    new_node = Node(value)
    if index == 0:
        new_node.next = node
        return new_node
    while index-1:
        node = node.next
        index -= 1
    tmp = node.next
    node.next = new_node
    new_node.next = tmp
    return head
```

### I. Стек - Max    

Нужно реализовать класс StackMax, который поддерживает операцию определения   
максимума среди всех элементов в стеке. Класс также должен поддерживать все операции,    
реализованные в классе Stack, из урока. При этом в классе StackMax может быть   
реализовано не более трёх методов.   

Стек может содержать только данные типов, поддерживающих операцию сравнения.    
Иначе операция поиска максимума будет некорректной.   

Формат ввода   
В первой строке записано одно число n - количество команд. n не превосходит 1000.    
В следующих n строках идут команды. Команды могут быть следующих видов:   

push x - добавить число x в стек   

pop - удалить число с вершины стека   

get_max - напечатать максимальное число в стеке   

Если стек пуст при вызове команды get_max нужно напечатать None, для команды pop - error.   

Формат вывода   
Для каждой команды get_max напечатайте результат её выполнения. Если стек пустой,    
для команды get_max напечатайте None. Если происходит удаление из пустого стека - напечатайте error.   

Ввод   
8   
get_max   
push 7   
pop   
push -2   
push -1   
pop   
get_max

Вывод:   
None   
-2   
-2   

Ввод:  
7   
get_max   
pop   
pop   
pop   
push 10   
get_max   
push -9   

Вывод:   
None   
error   
error   
error   
10   


```python
class StackMax:    
    def __init__(self):
            self.items = []
            self.max = []
            
    def isEmpty(self):
        return self.items == []
 
    def push(self, item):
        
        if len(self.items) == 0:
            self.max.append(int(item))
        elif int(item) > self.max[len(self.items)-1]:
            self.max.append(int(item))
        else:
            self.max.append(self.max[len(self.items)-1])
        self.items.append(item)
        

    def peek(self):
        return self.items[len(self.items)-1]

    
def pop(stack):
    if stack.isEmpty():
        return 'error'
    stack.max.pop()
    return stack.items.pop()

def size(stack):
    return len(stack.items)

def get_max(stack):
    if stack.isEmpty():
        return 'None'
    return stack.max[size(stack)-1]
```

```python
s = StackMax()
s.push(-1)
s.push(10)
s.push(4)

get_max(s)
```

```python
n = int(input())
s = StackMax()
res = []

for i in range(n):
    command = input().split()
    if len(command) == 2:
        s.push(command[1])
    if command[0] == 'pop':
        if pop(s) == 'error':
            res.append('error')
        
    if command[0] == 'get_max':
        res.append(get_max(s))
for x in res:
    print(x)
```

###  K. Уникальный Стек     

Реализуйте класс StackSet, который хранит только уникальные элементы.    
При этом операция добавления элемента в стек должна выполняться за O(1).

Далее идут команды по одной на строке. Команды могут быть следующих видов:

```python
from collections import defaultdict
class StackSet:    
    
    def __init__(self):
        self.items = []
        self.dict = defaultdict(int)
            
    def isEmpty(self):
        return self.items == []
 
    def push(self, item):
        if self.dict[item] == 0:
            self.items.append(item)
            self.dict[item] += 1
        
    def peek(self):
        if self.isEmpty():
            return 'error'
        return self.items[len(self.items)-1]
 
    

def pop(stack):
    if stack.isEmpty():
        return 'error'
    stack.dict[stack.peek()] -= 1
    return stack.items.pop()

def size(stack):
    return len(stack.items)

n = int(input())
s = StackSet()
res = []

for i in range(n):
    command = input().split()
    
    if len(command) == 2:
        s.push(command[1])
        
    if command[0] == 'pop':
        if pop(s) == 'error':
            res.append('error')
            
    if command[0] == 'peek':
        res.append(s.peek())
        
    if command[0] == 'size':
        res.append(size(s))
        
        
for x in res:
    print(x)
```

head — индекс, по которому нужно извлекать элемент, если очередь не пустая;   
tail — индекс, по которому нужно добавлять элемент, если в очереди есть место;

```python
class Queue:
    def __init__(self, n):
        self.queue = [None for _ in range(n)] 
        self.max_n = n
        self.head = 0
        self.tail = 0
        self.size = 0
        
    def is_empty(self):
        return self.size == 0

    def put(self, x):
        if self.size != self.max_n:
            self.queue[self.tail] = x
            self.tail = (self.tail + 1) % self.max_n
            self.size += 1

    def get(self):
        if self.is_empty():
            return None
        x = self.queue[self.head]
        self.queue[self.head] = None
        self.head = (self.head + 1) % self.max_n
        self.size -= 1
        return x
```

```python
class ListQueue:
    class Node:
        def __init__(self, value = None, next = None):
            self.value = value
            self.next = next

        def __str__(self):
            return self.value
        
    def __init__(self):
        #self.queue = self.Node() 
        self.head = self.Node()
        self.tail = self.Node()
        self.size = 0
        
    def is_empty(self):
        return self.size == 0

    def put(self, x):
        if self.size == 0:
            self.head = self.Node(value=x)
            self.tail = self.head
        else:
            self.tail.next = self.Node(value=x)
            self.tail.next.next = self.head
            self.tail = self.tail.next
        self.size += 1

    def get(self):
        if self.is_empty():
            return 'error'
        if self.size == 1:
            x = self.head
            self.head = self.Node()
            self.tail = self.Node()
            self.size -= 1
            return x
        if self.size == 2:
            x = self.head
            self.head = self.tail
            self.size -= 1
            return x
            
        x = self.head
        self.head = self.tail.next.next
        self.tail.next = self.head
        self.size -= 1
        return x
    
    def __str__(self):
        return self.value
```

```python
n = int(input())
s = ListQueue()
res = []

for i in range(n):
    command = input().split()
    
    if len(command) == 2:
        a = s.put(command[1])
        if a == 'error':
            res.append(a)
        
    if command[0] == 'get':
        res.append(s.get())
            
        
    if command[0] == 'size':
        res.append(s.size)
        
for x in res:
    print(x)
```

```python
from collections import defaultdict
class Dec:
    def __init__(self, m):
        self.dec = defaultdict(int) 
        self.max_m = m
        self.head = 0
        self.tail = 0
        self.size = 0

        
    def is_empty(self):
        return self.size == 0

    def push_back(self, x):
        if self.size < self.max_m:
            self.tail -= 1
            self.dec[self.tail] = x
            self.size += 1
            if self.size == 1:
                self.head = self.tail 
        else:
            return 'error'
            
    def push_front(self,x):
        if self.size < self.max_m:
            self.head += 1
            self.dec[self.head] = x
            self.size += 1
            if self.size == 1:
                self.tail = self.head
        else:
            return 'error'
        
        
    def pop_back(self):
        #print(self.size)
        if self.is_empty():
            return 'error'
        x = self.dec.pop(self.tail)
        self.size -= 1
        if self.size == 1:
            self.tail = self.head
        else:
            self.tail += 1
        return x
            


    def pop_front(self):
        if self.is_empty():
            return 'error'
        x = self.dec.pop(self.head)
        self.size -= 1
        if self.size == 1:
            self.head = self.tail 
        else:
            self.head -= 1
        return x
```

```python
n = int(input())
m = int(input())
s= Dec(m)

res = []

for i in range(n):
    command = input().split()
    
    
    if command[0] == 'push_front':
        a = s.push_front(command[1])
        if a == 'error':
            res.append(a)
            
    if command[0] == 'push_back':
        a = s.push_back(command[1])
        if a == 'error':
            res.append(a)
        
    if command[0] == 'pop_front':
        res.append(s.pop_front())
            
        
    if command[0] == 'pop_back':
        res.append(s.pop_back())
        
        
for x in res:
    print(x)
```
