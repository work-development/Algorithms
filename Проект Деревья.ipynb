### A. Кондратиева пирамида   


В Трешландии проводились ежегодные соревнования по заезду на барсучьих упряжках.    
Нужно обработать результаты забегов. Данные представлены в формате:    
имя участника - список заработанных очков      
Нужно отсортировать данные по такому принципу:                                                         
Если сумма набранных очков участника A больше, чем у участника В, то А должен идти в списке раньше.    
При подсчете суммы нужно учитывать только положительные значения очков.             
Если суммы равны, то первым должен идти участник, чье имя лексикографически меньше.        
При совпадении имен раньше должен идти участник, который во входных данных находится позже.    
При этом, если из имени участника можно составить английский вариант имени Кондратий    
(например, если имя nekondratiy или drkonxxxiyatt), то все правила отменяются.                  
Он должен идти в списке раньше. Если таких участников больше одного, то раньше должен идти тот,    
кто находится позже во входных данных. (Заработанные очки не рассматриваются).   

Формат ввода                                                          
В первой строке задано n - число участников. Оно не превосходит 10000.                   
В следующих n строках записана информация о каждом из участников в обозначенном формате.    
Имя состоит из строчных латинских букв. Длина имени не превосходит 100.                      
Длина списка набранных очков не больше 1000. Каждое из очков по модулю не превосходит 1000.

```python
n = int(input())

if n != 0:
    L = []
    for i in range(n):
        L.append([x if i == 0 else int(x) for i,x in enumerate(input().split())])
    for i in range(n):
        L[i] = [L[i],[sum([ x for x in L[i][1:] if x > 0]), *[-ord(x) for x in L[i][0]] ,i]]

    k = 0
    K = {'k':1,'o':1,'n':1,'d':1,'r':1,'a':1,'t':1,'i':1,'y':1}
    for i,x in enumerate(L):
        d = {}
        for y in x[0][0]:
            if y in K:
                d[y] = 1
        if d == K:    
            a = L.pop(i)
            L.insert(0,a)
            k += 1
    
    # Пирамидальной сортировка

    # Процедура для преобразования в двоичную кучу поддерева с корневым узлом i, 
    #что является индексом в arr[]. n - размер кучи

    def heapify(Q, m, i):
        smallest = i       # Initialize smallest as root
        l = 2 * i + 1    # left  = 2*i + 1
        r = 2 * i + 2    # right = 2*i + 2

      # Проверяем существует ли левый дочерний элемент меньший, чем корень
        if l < m and Q[i][1] > Q[l][1]:
            smallest = l

        # Проверяем существует ли правый дочерний элемент меньший, чем корень
        if r < m and Q[smallest][1] > Q[r][1]:
            smallest = r

        # Заменяем корень, если нужно
        if smallest != i:
            Q[i], Q[smallest] = Q[smallest], Q[i] 

            # Применяем heapify к корню.
            heapify(Q, m, smallest)


    # Основная функция для сортировки массива заданного размера
    def heapSort(Q):
        m = len(Q)
        # Построение min-heap.
        for i in range(m, -1, -1):
        #for i in range(k-1,n,1):
            heapify(Q, m, i)
            
        # Один за другим извлекаем элементы
        for i in range(m-1, 0, -1):
            Q[i], Q[0] = Q[0], Q[i] 
            heapify(Q, i, 0)

    Q = L[k:]
    heapSort(Q)
    Z = L[:k]
    Z.extend(Q)
    for x in Z:
        print(' '.join(list(map(str,x[0]))))
        
else:
    print('')
```

```python
n = int(input())
participants = []
for i in range(n):
    participants.append([x if i == 0 else int(x) for i,x in enumerate(input().split())])
for i in range(n):
    scores = participants[i][1:]
    i_name = participants[i][0]
    participants[i] = [participants[i],[sum([ score for score in scores if score > 0]), *[-ord(ch) for ch in i_name] ,i]]

kondratiys_number = 0
kondratiy = {'k','o','n','d','r','a','t','i','y'}
for i,participant in enumerate(participants):
    part_name = set()
    participant_name = participant[0][0]
    for ch in participant_name:
        if ch in kondratiy:
            part_name.add(ch)
    if kondratiy.issubset(part_name):    
        participant_with_kondratiy = participants.pop(i)
        participants.insert(0,participant_with_kondratiy)
        kondratiys_number += 1

# Пирамидальной сортировка

# Процедура для преобразования в двоичную кучу поддерева с корневым узлом i, 
#что является индексом в arr[]. n - размер кучи

def heapify(nokondratiys, m, i):
    smallest = i       # Initialize smallest as root
    l = 2 * i + 1    # left  = 2*i + 1
    r = 2 * i + 2    # right = 2*i + 2

  # Проверяем существует ли левый дочерний элемент меньший, чем корень
    if l < m and nokondratiys[i][1] > nokondratiys[l][1]:
        smallest = l

    # Проверяем существует ли правый дочерний элемент меньший, чем корень
    if r < m and nokondratiys[smallest][1] > nokondratiys[r][1]:
        smallest = r

    # Заменяем корень, если нужно
    if smallest != i:
        nokondratiys[i], nokondratiys[smallest] = nokondratiys[smallest], nokondratiys[i] 

        # Применяем heapify к корню.
        heapify(nokondratiys, m, smallest)


# Основная функция для сортировки массива заданного размера
def heapSort(nokondratiys):
    m = len(nokondratiys)
    # Построение min-heap.
    for i in reversed(list(range(m+1))):
        heapify(nokondratiys, m, i)

    # Один за другим извлекаем элементы
    for i in reversed(list(range(m))):
        nokondratiys[i], nokondratiys[0] = nokondratiys[0], nokondratiys[i] 
        heapify(nokondratiys, i, 0)

nokondratiys = participants[kondratiys_number:]
heapSort(nokondratiys)
ans = participants[:kondratiys_number]
ans.extend(nokondratiys)
for x in ans:
    print(' '.join(list(map(str,x[0]))))

```
